```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

titanic_data = pd.read_csv('./data/titanic.csv')

titanic_data.info()
titanic_data.isna().sum()
```

```python
titanic_data.drop(['Name', 'Ticket', 'Embarked'], axis=1, inplace=True)
#unecessary feature
```

```python
titanic_data['Age'].fillna(titanic_data['Age'].mean(),inplace=True)
titanic_data['Age'].isna().sum()

titanic_data
```

```python
titanic_data["Cabin"].isna().sum() / len(titanic_data) * 100

titanic_data.drop("Cabin", axis=1, inplace=True)
titanic_data
```

```python
titanic_data = pd.get_dummies(titanic_data, columns=["Sex"], drop_first=True, dtype=int)
titanic_data
```

```python
x=titanic_data[['PassengerId','Pclass','Age','SibSp','Parch','Fare','Sex_male']]
y=titanic_data['Survived']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
#Fit  Logistic Regression 
lr=LogisticRegression(max_iter=1000)
lr.fit(x_train,y_train)
```

```python
predict=lr.predict(x_test)
predict
```

```python
from sklearn.metrics import confusion_matrix
pd.DataFrame(confusion_matrix(y_test,predict),columns=['Predicted No','Predicted Yes'],index=['Actual No','Actual Yes'])
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predict))
```

```python
lr.score(x_test, y_test)
```
