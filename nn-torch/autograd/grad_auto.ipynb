```python
from tensor import MyTensor
from vis import draw_dot
import numpy as np
```

```python
x1 = MyTensor([[2.0, 2.0], [2.0, 2.0]], label="x1")
x2 = MyTensor([[0.0, 0.0], [0.0, 0.0]], label="x2")

# weights w1, w2
w1 = MyTensor([[-3.0, -3.0], [-3.0, -3.0]], label="w1")
w2 = MyTensor([[1.0, 1.0], [1.0, 1.0]], label="w2")

# bias of the neurons
b = MyTensor([[6.8813735870195432, 6.8813735870195432], [6.8813735870195432, 6.8813735870195432]], label="b")
```

```python
x1w1 = x1*w1; x1w1.label = 'x1*w1'
x2w2 = x2*w2; x2w2.label = 'x2*w2'
x1w1x2w2 = x1w1 + x2w2; x1w1x2w2.label = 'x1w1 + x2w2'
n = x1w1x2w2 + b; n.label = 'n'

# apply the hyperbolic tangent function to n
o = n.tanh(); o.label = 'o'
```

```python
draw_dot(o)
```

```python
o.backward()
```

```python
draw_dot(o)
```
